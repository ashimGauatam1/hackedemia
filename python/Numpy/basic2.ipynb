```python
import pandas as pd
n=int(input("number of students\n"))
Students={
    "S_name":[],
    "M_math":[],
    "s_science":[]
}
for i in range(0,n):
    name=input("enter name of students")
    Students["S_name"].append(name)
    math=int(input("enter marks in maths"))
    Students["M_math"].append(math)
    science=int(input("enter marks in science"))
    Students["s_science"].append(science)

data=pd.DataFrame(Students)
data.to_csv('student.csv',index=False)

plt.scatter(Students["M_math"],Students["s_science"])
```

```python
def stock_price_calc(x,m=4,c=10):
    return m*x+c
X=[i for i in range(1,13)]
Y_1=[stock_price_calc(i) for i in X]
Y_2=[stock_price_calc(i,m=-3,c=5) for i in X]
Y_3=[stock_price_calc(i,m=6,c=1) for i in X]
plt.plot(X,Y_1)
plt.plot(X,Y_2)
plt.plot(X,Y_3)
plt.legend(['2005','2025','2023'])
plt.xlabel('Months')
plt.ylabel('Stock Price')
```
