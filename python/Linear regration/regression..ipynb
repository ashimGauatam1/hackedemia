import dependencies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```


<h4 align="center">Reading data</h3>

```python
data_students=pd.read_csv("data/student_scores.csv")
display(data_students.head(6))
```

<h3 align="center">Visualing data</h3>

```python
X=np.array(data_students["Hours"])
Y=np.array(data_students["Scores"])

plt.scatter(X,Y)
plt.xlabel("Hours")
plt.ylabel("Scores")
```

<h3 align="center">regression</h3>

```python
def compute(x,y):
    n=len(x)
    x_mean=sum(x)/n
    y_mean=sum(y)/n
    num=0
    den=0
    for i in range(n):
        num+=(x[i]-x_mean)*(y[i]-y_mean)
        den+=(x[i]-x_mean)**2
    m=num/den
    c=y_mean-(m*x_mean)
    return m,c

x=np.array(data_students["Hours"])
y=np.array(data_students["Scores"])

m,c=compute(x,y)
print(m,c)
```

<h3 align="center"> Visualize Best Data

```python
y_var=[(m*i+c) for i in X]

plt.plot(X,Y)
plt.plot(x,np.array(y_var),color="red")
```

<h3 align="center">R square method


```python
for i in range(len(X)):
    num=(y[i]-y_var[i])**2
    deno=(y[i]-Y.mean())**2
    R=(1-num/deno)**2

print(R)
```
