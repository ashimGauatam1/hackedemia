<h3 align="center">Visualing data</h3>

```python
x=np.array(data_students["Hours"])
y=np.array(data_students["Scores"])

plt.scatter(x,y)
plt.xlabel("Hours")
plt.ylabel("Scores")
```

<h3 align="center">regression</h3>

```python
def compute(x,y):
    n=len(x)
    x_mean=sum(x)/n
    y_mean=sum(y)/n
    num=0
    den=0
    for i in range(n):
        num+=(x[i]-x_mean)*(y[i]-y_mean)
        den+=(x[i]-x_mean)**2
    m=num/den
    c=y_mean-(m*x_mean)
    return m,c

x=np.array(data_students["Hours"])
y=np.array(data_students["Scores"])

m,c=compute(x,y)
print(m,c)
```
